```python
import matplotlib.pyplot as plt
from matplotlib import font_manager

# y 轴数据（3 月每天的温度）
y = [
    11, 17, 16, 11, 12, 11, 12, 6, 6, 7,
    8, 9, 12, 15, 14, 17, 18, 21, 16, 17,
    20, 14, 15, 15, 15, 19, 21, 22, 22, 22,
    23
]

# x 轴（1~31 日）
x = range(1, 32)

# 设置图形大小
plt.figure(figsize=(20, 8), dpi=80)

# 绘制散点图
plt.scatter(x, y, label='3月份')

# 设置中文字体（Windows）
my_font = font_manager.FontProperties(
    fname=r'C:\Windows\Fonts\simsun.ttc',
    size=12
)

# x 轴刻度标签
xticks_labels = [f'3月{i}日' for i in x]

# 每隔 3 天显示一个刻度
plt.xticks(x[::3], xticks_labels[::3],
           fontproperties=my_font, rotation=45)

# 坐标轴标签
plt.xlabel('日期', fontproperties=my_font)
plt.ylabel('温度', fontproperties=my_font)

# 图例
plt.legend(prop=my_font)

# 显示图像
plt.show()
```
